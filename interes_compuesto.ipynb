Se elabora un modelo que calcule la el interes compuesto para determinada inversion a traves de los años teniendo en cuenta el interes anual la inversion mensual, la inversion inicial y la duracion en años  

```python
import matplotlib.pyplot as plt

def calcular_interes_compuesto(inversion_inicial, inversion_mensual, tasa_interes_anual, años):
    tasa_interes_mensual = (1 + tasa_interes_anual) ** (1/12) - 1
    inversiones = [inversion_inicial]
    for año in range(1, años + 1):
        inversion_nueva = inversiones[-1]
        for mes in range(12):
            interes = inversion_nueva * tasa_interes_mensual
            inversion_nueva += inversion_mensual
            inversion_nueva += interes
        inversiones.append(inversion_nueva)
    return inversiones

inversion_inicial = 100
inversion_mensual = 100
tasa_interes_anual = 0.1121635  # Tasa de interés anual en porcentaje
años = 30  # Duración en años

inversiones = calcular_interes_compuesto(inversion_inicial, inversion_mensual, tasa_interes_anual, años)

# Se Converte el rango de años a meses
rango_años = range(0, años + 1)

plt.plot(rango_años, inversiones)
plt.xlabel('Año')
plt.ylabel('Inversión acumulada')
plt.title('Crecimiento de la inversión con interés compuesto (anual)')
plt.grid(True)
plt.show()
```
